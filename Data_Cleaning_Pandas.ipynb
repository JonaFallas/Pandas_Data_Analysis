```python
import pandas as pd
```

```python
#Reading a xlsl file with a list of customers that need to be called
df = pd.read_excel('Customer Call List.xlsx')
df
```

```python
#Dropping duplicates from the list

df=df.drop_duplicates()
df
```

```python
#Deleting a column that is not necessary 

df= df.drop(columns= 'Not_Useful_Column')
df
```

```python
# Removing NA,NaN, N/a and special characters ('./_') from Last_Name columm.

df['Last_Name'] = df['Last_Name'].apply(lambda x: '' if pd.isna(x) or pd.isnull(x) or str(x)=='N/a' else str(x))

df['Last_Name']= df['Last_Name'].str.strip('./_')

df
```

```python
#Concatenating columns to create a new Full Name column

df['Full_Name']= df['First_Name']+' '+df['Last_Name']
df
```

```python
#Dropping 'First_Name' and 'Last_Name' columns.

df = df.drop(columns = ['First_Name','Last_Name'], axis=1)
df
```

```python
#Removing NA and NAN values from the phone_Number column

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: '' if pd.isna(x) or pd.isnull(x) else str(x))
df
```

```python
#Noticed some "N/a" strings that were not removed.
#Replacing strings with 'N/a' for a blank space.

df['Phone_Number'] = df['Phone_Number'].str.replace('N/a','',regex=True)                                        
df
```

```python
#Standardizing phone numbers format

df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^a-zA-Z0-9]','', regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: '' if x=='' else x[0:3]+'-'+x[3:6]+'-'+x[6:10])
df
```

```python
#Adding underscore ("_") to "Paying Customer" column name

df = df.rename(columns={'Paying Customer': 'Paying_Customer'})
df
```

```python
#Working on NaN/Na from Paying_Customer and Do_Not_Contact columns

df['Paying_Customer'] = df['Paying_Customer'].apply(lambda x: '' if pd.isna(x) or pd.isnull(x) or str(x)=="N/a" else str(x))
df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: 'No' if pd.isna(x) or pd.isnull(x) or str(x)=="N/a" else str(x))
df
```

```python
#Standardizing entries from Paying_Customer and Do_Not_Contact column entries.
df['Paying_Customer']= df['Paying_Customer'].apply(lambda x: ('Yes' if str(x).startswith('Y') else ('No' if str(x).startswith('N') else "")))

#Adding "No" to customers that did not specify if they wanted to be called.
df['Do_Not_Contact']= df['Do_Not_Contact'].apply(lambda x: ('Yes' if str(x).startswith('Y') else ('No')))
df
```

```python
#Deleting from the list the customers that don't want to be called.

for x in df.index:
    if df.loc[x,'Do_Not_Contact'] == 'Yes':
        df.drop(x, inplace = True)
df
```

```python
#Creating a second data frame for customers whose phone number is missing.

Missing_Num_df = pd.DataFrame(columns =['CustomerID','Phone_Number','Address','Paying_Customer','Do_Not_Contact','Full_Name'])

for x in df.index:
    
    # Checking if the 'Phone_Number' column is empty
    if df.loc[x,'Phone_Number'] == '':
        
        #Adding those rows to a new dataframe 
        Missing_Num_df = Missing_Num_df._append(df.loc[x], ignore_index=True)
        
        #Droping those customers from the original dataframe
        df.drop(x, inplace=True)
        
Missing_Num_df
```

```python
df= df.reset_index(drop=True)
df
```

```python
#Saving the list as CSV file.

df.to_csv('To_Call_list.csv')
```
